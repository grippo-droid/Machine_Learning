## SMOTE(Synthetic Minority Oversampling Technique)
SMOTE (Synthetic Minority Over-sampling Technique) is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
```

This can make multiple classes and can make multiple datapoints as well

```python
X,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,
                   weights=[0.90],random_state=12)
```

* **n_samples = 1000**

➡️ Total number of data points (rows)

👉 Dataset will have 1000 samples  

* **n_features = 2**

➡️ Number of input features (columns in X)

👉 Each sample has 2 features → useful for plotting in 2D   

*   **n_redundant = 0**

➡️ Number of redundant (duplicate/informationally useless) features

👉 Here, none — both features are meaningful

*  **n_clusters_per_class = 1**

➡️ Number of clusters within each class

👉 Each class forms one group in feature space

* weights = [0.90] ⚠️ Important

➡️ Controls class imbalance

👉 90% of samples belong to Class 0
👉 Remaining 10% belong to Class 1

So:

Class 0 ≈ 900 samples

Class 1 ≈ 100 samples

*  random_state = 12

➡️ Fixes randomness for reproducibility

👉 Same dataset generated every run


* Final Dataset Shape

X.shape → (1000, 2)

y.shape → (1000,)  



***For every row in X → there is one corresponding label in y.***

| Feature 1 | Feature 2 | Class (y) |
| --------- | --------- | --------- |
| 1.24      | -0.87     | 0         |
| -0.55     | 0.32      | 1         |
| 0.91      | -1.10     | 0         |


```python
import pandas as pd
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```
